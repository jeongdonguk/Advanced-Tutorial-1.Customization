## 사용자 지정 레이어 구현
 - \__init__ 은 여기서 모든 입력을 초기화하는 함수이다.
 - build는 입력 텐서의 형상을 알고 나머지 초기화 작업을 수행한다.
 - call은 순방향 계산을 수행한다.
 

```python
class mydenselayer(tf.keras.layers.Layer):
    def __init__(self, num_ouputs):
        super(mydenselayer, self).__init__()
        self.num_ouputs = num_ouputs
    
    def build(self, input_shape):
        self.kernel = self.add_weight('kernel',
                                     shape = [int(input_shape[-1]), self.num_ouputs]
        )

    def call(self, inputs):
        return tf.matmul(inputs, self.kernel)

layer = mydenselayer(10)
```

## 합성 레이어
### keras.Model 은 다음과 같은 메서드가 필요할 때 상속합니다.
 - Model.fit, Model.evaluate, Model.save
<br>아래의 예시는 Resnet 볼록입니다.


```python
class ResnetIdentityBlock(tf.keras.Model):
    def __init__(self, model_name, kernel_size, filters):
        super(ResnetIdentityBlock, self).__init__(name = model_name)
        filters1, filters2, filters3 = filters

        self.conv2a = tf.keras.layers.Conv2D(filters1, (1, 1))
        self.bn2a = tf.keras.layers.BatchNormalization()

        self.conv2b = tf.keras.layers.Conv2D(filters2, kernel_size, padding='same')
        self.bn2b = tf.keras.layers.BatchNormalization()

        self.conv2c = tf.keras.layers.Conv2D(filters3, (1, 1))
        self.bn2c = tf.keras.layers.BatchNormalization()

    def call(self, input_tensor, training=False):
        x = self.conv2a(input_tensor)
        x = self.bn2a(x, training = training)
        x = tf.nn.relu(x)

        x = self.conv2b(x)
        x = self.bn2b(x, training = training)
        x = tf.nn.relu(x)

        x = self.conv2c(x)
        x = self.bn2c(x, training = training)
        
        x += input_tensor
        return tf.nn.relu(x)

block = ResnetIdentityBlock('test', 1, [1, 2, 3])
```

```python
_ = block(tf.zeros([1,2,3,3]))
```

대부분의 경우 많은 레이어를 구성하는 모델은 단순히 한 레이어를 차례로 호출한다. 이것은 다음과 같이 간단한 코드로 표현이 가능해진다.

```python
my_seq = tf.keras.Sequential([
    tf.keras.layers.Conv2D(1, (1,1), input_shape=(None, None, 3)),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Conv2D(2, 1, padding='same'),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Conv2D(3, (1, 1)),
    tf.keras.layers.BatchNormalization()
])

my_seq(tf.zeros([1, 2 ,3 ,3]))
```
